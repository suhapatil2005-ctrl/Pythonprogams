# ***Array initialize methods:***

```python
#zeros array

zeros_array=np.zeros((2,3))
print("zeroes array:\n",zeros_array)


#ones arrray

one_array=np.ones((2,3))
print("ones array:\n",one_array)

#full arrray
full_array=np.full((3,4),5)
print("full array:\n",full_array)

#identity matrix

identity_array=np.eye(3)
print("identity matrix:\n",identity_array)


#empty array

em_array=np.empty((2))
print("empty array:\n",em_array)

#evenly spaced array
print("evenly spaced array:\n",np.arange(2,10,1))

#specific number of equaly spaced values between a range

print("evenly spaced:\n",np.linspace(1,20,5))

```

```python
#random array

print(np.random.randint(0, 100, size=(3,4), dtype=np.int64))

```
