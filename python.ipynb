```python
import random
print("Welcome!")
number=random.randrange(0,101)

while True:
    guess=int(input("guess a number between 1 to 100:"))
    print("your guess",guess)


    if guess<1 or guess>100:
        print("Your guess is out of range of 1 to 100.")
        continue

    if guess == number:
        print("you guess correctly.")
        break

    elif guess > number:
        difference=guess-number
        print("Your guess is greater than number by",difference)


    elif guess<number:
        difference=number-guess
        print("Your guess is less than number by",difference)

```
